```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import shap
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# --- 1. SIMULATING YOUR DATASET STRUCTURE ---
# Let's assume 4 features with different "directions" of association
np.random.seed(42)
N = 400
X_sim = pd.DataFrame({
    'Income': np.random.normal(50, 15, N),
    'Age': np.random.normal(40, 10, N),
    'Debt_Level': np.random.normal(20, 10, N),
    'Credit_Score': np.random.normal(600, 100, N)
})

# Simulate a target: Higher Income/Credit = Class 1, Higher Debt = Class 0
y_sim = (0.5 * X_sim['Income'] + 0.2 * X_sim['Credit_Score'] - 0.8 * X_sim['Debt_Level'] > 40).astype(int)

# Splitting and Scaling (Essential for KNN)
x_train, x_test, y_train, y_test = train_test_split(X_sim, y_sim, test_size=0.2, random_state=42)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# --- 2. YOUR ORIGINAL KNN MODEL ---
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train_scaled, y_train)

# --- 3. SHAP EXPLANATION (DIRECTION OF ASSOCIATION) ---
print("Calculating SHAP values (using KMeans summary to speed up KNN)...")

# Summarize the background data using 10 k-means clusters. 
# This represents the 'average' behavior and makes the math much faster.
background_summary = shap.kmeans(x_train_scaled, 10)

# Initialize KernelExplainer with the probability function
explainer = shap.KernelExplainer(knn.predict_proba, background_summary)

# Calculate SHAP values for the test set
# (We'll use the first 100 samples for the simulation)
shap_values = explainer.shap_values(x_test_scaled[:100])

# --- 4. VISUALIZATION ---
plt.figure(figsize=(10, 6))

# If binary classification, shap_values is a list [negative_class, positive_class]
# We plot index [1] to see what DRIVES the prediction toward the positive outcome
shap.summary_plot(
    shap_values[1], 
    x_test_scaled[:100], 
    feature_names=X_sim.columns,
    plot_type="dot"
)
```
